# Hb 12 Spectral Mosaics

Some code to make spectral maps from a datacube. Program loads three data files in the form of a data cube for three pointings, separate observations, of a planetary nebula. The coordinates of the datacubes are x: Celestial E/W (R.A), y: Celestial N/S (Dec.), z: wavlength. The data is shifted and stacked to form a large datacube of the complete observation of the planetary nebula. Then, a spectral map is extracted at a specific wavelength in the datacube and output as a .png figure.

```python
import aplpy
import numpy as np
import pyfits
import os
import matplotlib
matplotlib.use('Agg')
from pyfits import getdata
from matplotlib import pyplot as plt
from matplotlib import mlab
from matplotlib import rc
from pylab import *
```

```python
# get system arguments
postype = sys.argv[1]
imtype = sys.argv[2]
lineLabel = sys.argv[3]
getWave = sys.argv[4]
```

```python
# Define additional variables
fpathOut='{output file path}'
linetype = ''
if lineLabel == '[Fe II]': linetype = 'FeII'
if lineLabel == 'H$_2$': linetype = 'H2'
if lineLabel == 'Br$\gamma$': linetype = 'BrGamma'
if lineLabel == 'He I': linetype = 'HeI'

dwave = 3
padsize = 10
getWaveLabel = float(getWave)*10000.
```

```python
# Load fits files
if postype == 'fh':
    fpathpos1 = '{data file path}/Hb12-H-pos1_noSS/'
    if imtype == 'source':
        fitsfile1 = 'ctfbrnN20100925S0235.fits'
    if imtype == 'sky':
        fitsfile1 = 'ctfbrnN20100925S0236.fits'
    if imtype == 'sub':
        fitsfile1 = 'cubecombine.fits'
    data1, header1 = getdata(fpathpos1+fitsfile1, header=True)
    pad = np.zeros((header1['NAXIS3'], header1['NAXIS2'], 25))
    data1 = np.dstack((pad, data1))
    data1 = data1[:,0:45,:]
    
    fpathpos2 = '{data file path}/Hb12-H-pos2_noSS/'
    if imtype == 'source':
        fitsfile2 = 'ctfbrnN20100925S0262.fits'
    if imtype == 'sky':
        fitsfile2 = 'ctfbrnN20100925S0263.fits'
    if imtype == 'sub':
        fitsfile2 = 'cubecombine.fits'
    data2, header2 = getdata(fpathpos2+fitsfile2, header=True)
    pad = np.zeros((header2['NAXIS3'], header2['NAXIS2'], 25))
    data2 = np.dstack((data2, pad))
    
    fpathpos3 = '{data file path}/Hb12-H-pos3_noSS/'
    if imtype == 'source':
        fitsfile3 = 'ctfbrnN20100925S0288.fits'
    if imtype == 'sky':
        fitsfile3 = 'ctfbrnN20100925S0289.fits'
    if imtype == 'sub':
        fitsfile3 = 'cubecombine.fits'
    data3, header3 = getdata(fpathpos3+fitsfile3, header=True)
    pad1 = np.zeros((header3['NAXIS3'], header3['NAXIS2'], 11))
    pad2 = np.zeros((header3['NAXIS3'], header3['NAXIS2'], 14))
    data3 = np.dstack((pad1, data3, pad2))
    pad3 = np.zeros((data3.shape[0], 16, data3.shape[2]))
    data3 = np.hstack((data3, pad3))
    
if postype == 'fk':
    fpathpos1 = '{data file path}/Hb12-F1-K_noSS/'
    if imtype == 'source':
        fitsfile1 = 'ctfbrnN20100925S0230.fits'
    if imtype == 'sky':
        fitsfile1 = 'ctfbrnN20100925S0231.fits'
    if imtype == 'sub':
        fitsfile1 = 'cubecombine.fits'
    data1, header1 = getdata(fpathpos1+fitsfile1, header=True)
    pad = np.zeros((header1['NAXIS3'], header1['NAXIS2'], 26))
    data1 = np.dstack((pad, data1))
    data1 = data1[:,0:45,:]
    
    fpathpos2 = '{data file path}/Hb12-F2-K_noSS/'
    if imtype == 'source':
        fitsfile2 = 'ctfbrnN20100925S0257.fits'
    if imtype == 'sky':
        fitsfile2 = 'ctfbrnN20100925S0258.fits'
    if imtype == 'sub':
        fitsfile2 = 'cubecombine.fits'
    data2, header2 = getdata(fpathpos2+fitsfile2, header=True)
    pad = np.zeros((header2['NAXIS3'], header2['NAXIS2'], 26))
    data2 = np.dstack((data2, pad))
    
    fpathpos3 = '{data file path}/Hb12-K-pos3_noSS/'
    if imtype == 'source':
        fitsfile3 = 'ctfbrnN20100925S0283.fits'
    if imtype == 'sky':
        fitsfile3 = 'ctfbrnN20100925S0284.fits'
    if imtype == 'sub':
        fitsfile3 = 'cubecombine.fits'
    data3, header3 = getdata(fpathpos3+fitsfile3, header=True)
    pad1 = np.zeros((header3['NAXIS3'], header3['NAXIS2'], 7))
    pad2 = np.zeros((header3['NAXIS3'], header3['NAXIS2'], 19))
    data3 = np.dstack((pad1, data3, pad2))
    pad3 = np.zeros((data3.shape[0], 15, data3.shape[2]))
    data3 = np.hstack((data3, pad3))
```

```python
# Square central position for display purposes
data1 = np.square(data1)
```

```python
#stack fits files to make large data cube for all three points
dataStack = np.hstack((data3, data1, data2))    

headerOut = header1
headerOut['NAXIS1'] = dataStack.shape[2]
headerOut['NAXIS2'] = dataStack.shape[1]
headerOut['NAXIS3'] = dataStack.shape[0]

pyfits.writeto(fpathOut+'out.fits', dataStack, headerOut, clobber=True)

xstep = header1['CD1_1']
xsize = headerOut['NAXIS1']
ysize = headerOut['NAXIS2']
lstep = header1['CD3_3']
lmin = header1['CRVAL3']
lsize = header1['NAXIS3']
lmax = lmin+lstep*lsize

getWave = float(getWave)

wave = np.arange(lmin, lmax, (lmax - lmin)/lsize)/10000.

pix = np.argmin(np.absolute(wave - getWave))

dataCube, headerCube = getdata(fpathOut+'out.fits', header=True)

dataCubeCut = np.sum(dataCube[pix-dwave:pix+dwave,:,:], axis=0)
```

```python
#update header
headerCut = headerOut
headerCut['NAXIS1'] = dataCubeCut.shape[0]
headerCut['NAXIS2'] = dataCubeCut.shape[1]
headerCut['WCSDIM'] = 2
del headerCut['CD3_3']
del headerCut['NAXIS3']
del headerCut['CRVAL3']
del headerCut['WAT3_001']
del headerCut['CRPIX3']
del headerCut['CTYPE3']
del headerCut['LTM3_3']

pyfits.writeto(fpathOut+'out2.fits', dataCubeCut, headerCut, clobber=True)

z0 = 80.
z1 = 99.99999
```

```python
# Make figure of nebula at specific wavelength in datacube
fig = aplpy.FITSFigure(fpathOut+'out2.fits', figsize=(7,7))
fig.add_label((xsize/2)*xstep, ysize*xstep+0.2, lineLabel+' ('+str(np.round_(getWave,6))+' $\mu$m), '+imtype)
fig.add_label(4.6, 0.8, 'z0 = '+str(z0), horizontalalignment='left') # greyscale settings
fig.add_label(4.6, 0.5, 'z1 = '+str(z1), horizontalalignment='left') # greyscale settings
fig.axis_labels.set_xtext('arcseconds')
fig.axis_labels.set_ytext('arcseconds')
fig.show_grayscale(stretch='sqrt', invert=True, pmin=z0, pmax=z1)
fig.show_contour(fpathOut+'out2.fits', levels=7, colors='blue')
fig.save(fpathOut+'Mosaics/'+linetype+'_'+str(int(getWave*10000))+'_core_'+imtype+'.png')
show()
```
